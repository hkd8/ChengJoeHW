```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "Resources/city_data.csv"
ride_data = "Resources/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

# Combine the data into a single dataset
merged_df_dupes = ride_df.merge(city_df, on = 'city')
merged_df_nodupes = merged_df_dupes.drop_duplicates(subset='city')
# Display the data table for preview
merged_df_nodupes.head()
```

## Bubble Plot of Ride Sharing Data

```python
#Grouping by cities; making a dupes/nodupes dataframe
city_grouped = merged_df_dupes.groupby('city')
city_grouped_nodupes = merged_df_nodupes.groupby('city')

#Finding my variables
average_fare = city_grouped['fare'].mean()
total_rides = city_grouped['ride_id'].count()
total_drivers = city_grouped_nodupes['driver_count'].mean()

#Variable of city type, note use of the original dataframe, NOT the city grouped dataframe. RESULTS WILL BE DIFFERENT.
city_type = merged_df_nodupes.set_index('city')['type']

#the master dataframe
master_df = pd.DataFrame({'Average Fare': average_fare,
                         'Total Rides': total_rides,
                         'Total Drivers':total_drivers,
                         'City Type': city_type})

#Using the master dataframe to seperate into city types
urban_df = master_df[master_df['City Type'] == 'Urban']
suburban_df = master_df[master_df['City Type'] == 'Suburban']
rural_df = master_df[master_df['City Type'] == 'Rural']

#Building each scatter plot, per city type.
#Note the *10 to make the scatter points more visible. 
#Note, first input is x, second input is y, third input are plot points. 
plt.scatter(rural_df['Total Rides'], rural_df['Average Fare'], s = rural_df['Total Drivers']*10, color = ['gold'], edgecolor = 'black', label = 'Rural', alpha = .75)
plt.scatter(suburban_df['Total Rides'], suburban_df['Average Fare'], s = suburban_df['Total Drivers']*10, color = ['lightskyblue'], edgecolor = 'black', label = 'Suburban', alpha = .75)
plt.scatter(urban_df['Total Rides'], urban_df['Average Fare'], s = urban_df['Total Drivers']*10, color = ['coral'], edgecolor = 'black', label = 'Urban', alpha = .75)

#Assigning grid lines to the graph
plt.grid()

#Labeling your graph. 
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

# Setting the legend variables
legend = plt.legend(fontsize = 8, title= "City Types") 
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

#Adding text notes to the graph
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save Figure
plt.savefig('Pyber Ride Sharing Data')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
master_type = merged_df_dupes.groupby(['type'])

total_fare_by_type = master_type['fare'].sum()
# Build Pie Chart
pie_labels = ['Urban', 'Suburban', 'Rural']
explode = (0, 0, 0.1)
colors = [ "gold","lightskyblue", "lightcoral"]
plt.pie(total_fare_by_type, explode=explode, labels=pie_labels, colors=colors, autopct="%1.1f%%", shadow=True,  startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig('Total Fare by City Type')
```

```python
# Show Figure
plt.show()
```
